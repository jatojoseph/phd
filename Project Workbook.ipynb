```python
# creating column for t_i and CST
df2['t_i'] = 0
df2['CST'] = 0

# determination of the max between t_i and CST
# In the ATCS analysis, X = max(t_i, CST)
X = []
for val in range(len(df2)):
    X.append(max(df2.iloc[val,15], df2.iloc[val, 16]))
df2['X'] = X

# In the ATCS analysis, Y = d_j - P_ij - X
Y = []
for val in range(len(df2)):
    Y.append(df2.iloc[val,3] - df2.iloc[val, 0] - df2.iloc[val, 17])
df2['Y'] = Y


# In the ATCS analysis, Z = max(Y,0)
Z = []
for val in range(len(df2)):
    Z.append(max(df2.iloc[val, 18], 0))
df2['Z'] = Z


K_1 = 4.7
K_2 = 0.48
P = df2['P_ij(M2)'].mean()
S_hj = 0
s = 0

# In the ATCS analysis, the formula used is Index = (w_j/P_ij)exp((-max(Y,0))/(K_1*P))
Index = []
for val in range(len(df2)):
    Index.append((df2.iloc[val, 2]/df2.iloc[val, 0])*exp((-df2.iloc[val, 19])/(P*K_1)))
df2['Index'] = Index
```
